```python
#IMPORTACION DE LIBERIAS

# Manejo de datos
import numpy as np
import pandas as pd

# Visualizacion
import matplotlib.pyplot as plt
import seaborn as sns

from sqlalchemy import create_engine
import os
from dotenv import load_dotenv
```

```python
# Cargar las credenciales deñ archivo .env
load_dotenv()

USER = os.getenv('POSTGRES_USER')
PASSWORD = os.getenv('POSTGRES_PASSWORD')
HOST = os.getenv('POSTGRES_HOST')
PORT = os.getenv('POSTGRES_PORT') 
DB = os.getenv('POSTGRES_DB')

# 2. Crear la conexión a PostgreSQL
engine = create_engine(f'postgresql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DB}')

print("Conexión a la base de datos establecida")
```

```python
# 1. Definir la consulta SQL
query = """
SELECT * FROM olist_orders_dataset 
LIMIT 10;
"""

# 2. Ejecutar la consulta y guardarla
df_orders = pd.read_sql(query, engine)

# 3. Visualizar 
df_orders.head()
```

```python


# Configurar el estilo visual (para que se vea más profesional)
sns.set_theme(style="whitegrid")

# Crear el lienzo usando Matplotlib
plt.figure(figsize=(10, 5))

# Dibujar el gráfico usando Seaborn (ejemplo: contar cuántos pedidos hay por estado)
sns.countplot(data=df_orders, x='order_status', palette='viridis')

# Añadir títulos y detalles
plt.title('Distribución del Estado de los Pedidos', fontsize=14)
plt.xlabel('Estado del Pedido')
plt.ylabel('Cantidad de Pedidos')

# Mostrar el gráfico
plt.show()
```
